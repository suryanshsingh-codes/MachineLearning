# *Model Creation*

Now i want to create a model
*   which will do the prediction
    * x variable  - input features: all except Prices
    * y variable  - output features: Prices  

**1. Creating the input and output variables**

```python
xInput = myData.drop(columns=['price'], axis=0) # all input features
#  axis =1 : columns, it is not working in vs code but works fine in jupyter notebook

yOutput = myData['price'] # all output features

```

```python
xInput.head()
```

```python
yOutput.head()
```

**2. Now i will encode all the categorical data into the numerical data**

* 1. Now from ONE HOT ENCODING

```python
# i will use one hot encoding
xInputoneHot=pd.get_dummies(xInput, columns=
                                             ['model','transmission','fuelType'],
                                            drop_first = True)

#  now converting these true false vallues into the 0 and 1
xInputoneHot=xInputoneHot.astype(int)

xInputoneHot
```

*   2. Now from LABEL ENCODING

```python
# import label encoder
from sklearn.preprocessing import LabelEncoder
# apply it
myLabelEncoder=LabelEncoder()
# do the label encoding of selected columns
myColumns= ['model', 'transmission', 'fuelType']
#  make the copy of xInputVaribles for label encoding
xInputLab=xInput.copy()

# THE ORIGINAL WORK OF LABEL ENCODING
for i in myColumns:
    # calling the label encoder function to use it
    le=LabelEncoder()
    # now fitting the values of string in those columns
    xInputLab[i]=le.fit_transform(xInputLab[i].astype(str))
   
xInputLab
```

# **3. Now Train test split**

1. *before this we will do STANDARD SCALING and bring all to same scale*
    1. year
    2. mileage
    3. tax
    4. mpg
2. *Then scalar fit of label encoding for better comparision*
3. *Create two models*
    1. using one hot encoding x and y 
    1. using label encoding x and y 
4. *Now Finally do the train test split*

1. *STANDARD SCALING and bring all to same scale*

```python
from sklearn.preprocessing import StandardScaler

# Numerical columns will go for standard scaling
numericalColumns = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
# initiallizing object to use its function
scaler = StandardScaler()
# ACTUALLY DOING STANDARD SCALING
xInputoneHot[numericalColumns] = scaler.fit_transform(xInputoneHot[numericalColumns])

xInputoneHot # now i can use this in model bcz it has been standard scaled
```


2. *NOW for label encoding*  
        *->  we will do SCALAR FIT*  
        *->  after its standard scaling has been done*

to bring all the feature on the same scale for better comparision





```python
xInputLab
```

```python
xInputLab[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg','engineSize']] = scaler.fit_transform(
                                                                                                  xInputLab[
                                                                                                         ['model', 'year', 'transmission', 'mileage', 
                                                                                                         'fuelType', 'tax', 'mpg','engineSize']
                                                                                                  ]
                                                                                                  )
                                                                                           
xInputLab
```

**3. Creating models**

```python
# import all libraries

from sklearn.model_selection import train_test_split # for train test split
from sklearn.linear_model import LinearRegression # for creating a linear regresssion model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score # for model evaluation
```

* 3.1 Using ONE HOT ENCODING

```python
#  1. train test split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(xInputoneHot, yOutput, test_size=0.33, random_state=42)
```

```python
# 2. creating the linear regression model
model = LinearRegression()
model.fit(Xtrain, Ytrain)
```

```python
#  3. Checking the predicion
yPred=model.predict(Xtest)
yPred
```

```python
Ytest
```

```python
#  4. comparing y Prediction and y Test: use r2 and adjusted r2
r2=r2_score(Ytest,yPred)
r2
```

```python
#   4. comparing y Prediction and y Test: use r2 and adjusted r2
n = Xtest.shape[0]
p = Xtest.shape[1]
adjustedr2 = 1 - ((1 - r2) * (n - 1)) / (n - p - 1)
print("Adjusted R² Score:", adjustedr2)
```

* 3.2 Using LABEL ENCODING

```python
#  1. train test split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(xInputLab, yOutput, test_size=0.33, random_state=42)
```
