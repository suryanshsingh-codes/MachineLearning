```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Create synthetic dataset with 500 samples, 5 features, 3 clusters
X, y = make_blobs(n_samples=500, n_features=5, centers=3, cluster_std=1.5, random_state=42)
```

```python
X
```

```python
# Standardize features (important before PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# Apply PCA to reduce dimensions from 5 → 2
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

```python
# Convert PCA output to DataFrame
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
df_pca['label'] = y   # Add original cluster labels
```

```python
# Visualize clusters in 2D using PCA components
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='label', palette='Set2')
```
