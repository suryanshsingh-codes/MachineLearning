# 2. *Data Cleaning*

```python
# 1. Drop the data which are null
myData.drop(["deck", "embark_town", "alive", "class", "who", "adult_male"], axis=1, inplace=True)
```

```python
#  Again check your data after cleaning
myData
```

```python
# information of data again check
myData.info()
```

```python
# 2. fill the null values with mean median and mode
myData["age"].fillna(myData["age"].mean(), inplace=True)

```

```python
# delete the rows which are null
myData.dropna(subset=["embarked"], inplace=True)
```

```python
#  again check the null values
myData.info()
```

# 3. *Label Encoding*

```python
# import the label encoder
from sklearn.preprocessing import LabelEncoder
myLabEncObject = LabelEncoder()
```

```python
myData.head()
```

**1. Sex, Embarked Values are only in numeric**
* Label encode it

```python
myData['sex'] = myLabEncObject.fit_transform(myData['sex'])
myData["embarked"] = myLabEncObject.fit_transform(myData["embarked"]) # S=2, C=0, Q=1
```

```python
myData
```

* Now convert rest value of ALONE true false into 1 and 0

```python
myData=myData.astype(int)
```

```python
myData
```

# 4. *Logistic Regression and its Model*

**1. Selecting the variables**
* 1. x=Input Variable- all except survived
* 2. y=Output Variable- those who survived

```python
xInput = myData.drop("survived", axis=1)
yOutput = myData["survived"]
```

**2. Train Test Split**

```python
from sklearn.model_selection import train_test_split

xTrain, xTest, yTrain, yTest = train_test_split(xInput, yOutput, test_size=0.2, random_state=42)
```

**3. Logistic Model Creation**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

# fit the modelto train it
model.fit(xTrain,yTrain)
```

**4. Prediction: The predicted y data**

```python
yPred = model.predict(xTest)
yPred
```

**4. Y-Test : The real data**

```python
yTest
```

# **Model Evaluation**

**Here we will use Confusion matrix**  
**as same we were using r2 and adjusted r2 in Linear regression**

```python
from sklearn.metrics import accuracy_score, confusion_matrix , classification_report
```

**1. accuracy score**

```python
accuracy_score(yTest,yPred)
```

**2. Confusion matrix**

```python
confusion_matrix(yTest,yPred)
```

**3. Now print classification report which will contain all the reports**

```python
print(classification_report(yTest,yPred))
```

# 5. *KNN Algorithm and its model*

Before using KNN, we will need to do feature scaling because, here AGE and FARE values are higher which may impact the prediciton

**1. Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
# initializing the object to use its functions
myScaler = StandardScaler()
```

```python
#  now scale the trained data and testing data
xTrainScaled = myScaler.fit_transform(xTrain)
xTestScaled = myScaler.fit_transform(xTest)
```

**2. Now import KNN Algo**

```python
from sklearn.neighbors import KNeighborsClassifier

# now start the object and put k value =5
knnModel = KNeighborsClassifier(n_neighbors= 5)
#  now train the model on xTrainScaled and yTrain
knnModel.fit(xTrainScaled,yTrain)
```

**3. Y Predcition**

```python
# now see the knn prediction
yPredKnn = knnModel.predict(xTestScaled)
```

```python
yPredKnn
```

**4. ACCURACY SCORE**

```python
accuracy_score(yTest,yPredKnn)
```

**5. CONFUSION MATRIX**

```python
confusion_matrix(yTest,yPredKnn)
```

**6. Classification Report**

```python
print(classification_report(yTest,yPredKnn))
```
